```python
def show_image_3(image, image_label):
    # Convert image to 8-bit unsigned integer type for display
    image_display = (image * 255).astype(np.uint8)

    # If the image has more than one channel, convert it to RGB
    if image_display.shape[-1] == 1:
        image_display = cv2.cvtColor(image_display, cv2.COLOR_GRAY2RGB)
    elif image_display.shape[-1] == 3:
        image_display = cv2.cvtColor(image_display, cv2.COLOR_BGR2RGB)

    plt.imshow(image_display)
    plt.title(image_label)
    plt.show()
```

```python
def generate_random_palette(num_colors):
    # Asegurarse de que num_colors sea al menos 1 para evitar la divisi√≥n por cero
    if num_colors < 1:
        raise ValueError("num_colors debe ser al menos 1")

    palette = np.zeros((num_colors, 3), dtype=np.uint8)
    for i in range(num_colors):
        # Generar valores aleatorios para los canales R, G y B
        red_value = np.random.randint(0, 256)
        green_value = np.random.randint(0, 256)
        blue_value = np.random.randint(0, 256)

        palette[i] = [red_value, green_value, blue_value]

    return palette
```

```python
def pseudocolor_intensity_slice(image, num_levels):
    if len(image.shape) == 2:
        # Grayscale image
        rows, cols = image.shape
    elif len(image.shape) == 3:
        # Convertir la imagen a escala de grises
        image = rgb2gray(image)
        rows, cols = image.shape
    else:
        raise ValueError("Unsupported image shape")

    rgb_image = np.zeros((rows, cols, 3), dtype=np.uint8)
    palette = generate_random_palette(num_levels)
    # Normalizar la imagen para que est√© en el rango [0, 1]
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Asignar colores utilizando np.digitize
    bin_indices = np.digitize(normalized_image, np.linspace(0, 1, num_levels)) - 1
    rgb_image = palette[bin_indices]

    return rgb_image
```

```python
num_levels = 200

plt.figure(figsize=(20, 20))
gray_image = plt.imread("medtest.png")
# Obtener la imagen pseudocolor
pseudocolor_image = pseudocolor_intensity_slice(gray_image, num_levels)
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen en Escala de Grises')

plt.subplot(1, 2, 2)
plt.imshow(pseudocolor_image)
plt.title('Imagen Pseudocolor')

plt.show()
plt.subplots_adjust(left=0.1, right=0.6, bottom=0.1, top=0.3, wspace=0.2, hspace=0.4)
```
